# Import libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```


# Creating a Synthetic DataSet



```python
np.random.seed() # For Reproducibility
x = np.random.rand(100,3)
y = x[:,0]* 150000 + x[:,1]* 100000 + x[:,2] * 500000 + 500000
```

# Converting to a Dataframe

```python
data = pd.DataFrame(x, columns=['Size','Bedroom','Age'])
data['price'] = y

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)

data.head()
```

# Initializing the Model

```python
model = LinearRegression()
```

# Fitting The Model

```python
model.fit(x_train, y_train)
```

# Making Predition

```python
y_predict = model.predict(x_test)
#print (x_test)
#print (y_predict)
```

# Calculating the Mean Squarded Error

```python
mse = mean_squared_error(y_test, y_predict)
print(f"Meam squared error:{mse}")
```

# Predicting the Price for A new House

```python
new_house = [[0.5,2,10]]
predicted_price = model.predict(new_house)
print(f"Predicted : {predicted_price[0]}")
```
