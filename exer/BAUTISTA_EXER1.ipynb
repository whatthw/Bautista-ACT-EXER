1. importing Libraies

```python
import pandas as pd #Data Manipulation
import matplotlib.pyplot as plt # Plot / Visualization
```

2.Loading Dataset

```python
data = pd.read_csv('sales_data_2.csv')
```

3. Data loading and InitialData


```python
statistic = data.describe()
```

4. Data Manipulation

```python
product_revenue = data.groupby('Product')['Revenue'].sum()
highest_revenue_product = product_revenue.idxmax()
average_quantity_per_day = data['Quantity Sold'].mean()
```

5. Data Visualization

```python
product_revenue.plot(kind='bar')
plt.xlabel('Product')
plt.ylabel('revenue')
plt.title('Revenue By Product')
plt.xticks(rotation=55)
plt.tight_layout()
plt.show()
```

6. Analysis report

```python
report = """
Analysis Report
1. Data Loading and Initialdata :
    Basic statistic of Dataset:
    {}
2. Data Manipulation:
    Total Revenue for each product:
    {}
    product with the highest revenue: {}
    Average quantity sold per day {:.2f}
3. Data visualization:
    Bar char display revenue for each product is attached
4. Insights:
    The product {} generate the hightest revenue
    On average , {:.2f} units are sold per day.

""". format(statistic,product_revenue, highest_revenue_product, average_quantity_per_day,
            highest_revenue_product,average_quantity_per_day)

print(report)
```

7. Saving report to file

```python
with open('report.txt','w')as f:
  f.write(report)
```
