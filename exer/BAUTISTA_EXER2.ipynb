1. Creating Dataset


```python
import pandas as pd

data = {
    'student_id': range(1,16),
    'gender': ['male','male','female','male','male','male','female','male','male','female','male','female','male','female','male'],
    'test_date': pd.date_range(start = '2024-2-14', periods=15),
    'study_hours':[5,4,6,4,5,6,7,8,3,4,3,3,4,6,8],
    'test_score':[65,67,47,86,48,69,66,47,47,32,58,92,90,36,37],
    'attendance':['present','absent','present','present','absent','present','present','absent','present','present','absent','present','present','absent','presnet']

}

```

```python
df = pd.DataFrame(data)

df.to_csv('student_performance.csv', index=False)
```

2.Data Loading and Initial Examinition

```python
df = pd.read_csv('student_performance.csv')

print(df.head())

print(df.isnull().sum())
```

exploratory data analiysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['test_score', 'study_hours']].describe()

#histogram
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['test_score'],bins = 15, kde = True)
plt.title('Distributions of test score')


```

Data Visualization

```python
plt.figure(figsize=(6,4))
sne.countplot(data=df,x='gender')
plt.title('Gender distribution')


plt.figure(figsize=(24,8))
sne.lineplot(data=df,x='test_date',y='test_score')
plt.title('Gender distribution')

correlation_matrix = df[['study_hours','test_score','attendance']].corr()
plt.figure(figsize=(24,6))
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('correlation Heatmap')
```
