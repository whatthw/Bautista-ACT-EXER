1. Import Liblareis

```python
import pandas as pd # Data Manipulation
import numpy as np # Mathematical Operation
import matplotlib.pyplot as plt # Data Visualization
from sklearn.linear_model import LinearRegression # Training Data Modeling
```

2. Load Data

```python
data = pd.read_csv('Lesson_4_student_grades_dataset.csv')
```

3. Visualizing the Data

```python
plt.scatter(data['Attendance_Percentage'],data['Grades'])
plt.xlabel('Independent var (Attendance_Percentage)')
plt.ylabel('Dependent var (Grades)')
plt.title('Grades')
plt.show()
```

Prepering Model

```python
x = data [['Attendance_Percentage']]
y = data [['Grades']]
```

Fitting the Linear Regression Model:

```python
model = LinearRegression()
model.fit(x,y)
```

Getting Model Coefficeints:

```python
slope = model.coef_[0]
intercept = model.intercept_
print("Slope" , slope)
print(" intercept" , intercept)
```

Visualizing the Linear Regression Line:

```python
plt.scatter(x,y)
plt.plot(x,model.predict(x), color= 'blue')
plt.xlabel('Independent var (x)')
plt.ylabel('Dependent var (y)')
plt.title('Linear Regression')
plt.show()
```

```python
new_x = [[np.random.randint(100,250)]]
predicted_y = model.predict(new_x)
print ("Value of x:", new_x)
print("Predict y to new_x:", predicted_y )
```
