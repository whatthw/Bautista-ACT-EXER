DataLoading

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Data Loading
df = pd.read_csv('/full/path/to/your/hardwareStore.csv')


# 2. Initial Data Analysis
# Basic statistics
basic_stats = df.describe()
print("Basic Statistics:")
print(basic_stats)

# Unique categories and counts
category_counts = df['Category'].value_counts()
print("\nUnique Categories and Counts:")
print(category_counts)

# 3. Category Analysis
# Bar chart for distribution of products across categories
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Products Across Categories')
plt.xlabel('Categories')
plt.ylabel('Number of Products')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Category with the highest number of products
highest_category = category_counts.idxmax()
print("\nCategory with the highest number of products:", highest_category)

# 4. Cost and Price Analysis
# Average standard cost and list price
avg_standard_cost = df['Standard Cost'].mean()
avg_list_price = df['List Price'].mean()
print("\nAverage Standard Cost:", avg_standard_cost)
print("Average List Price:", avg_list_price)

# Product with the highest standard cost and list price
highest_standard_cost_product = df.loc[df['Standard Cost'].idxmax()]
highest_list_price_product = df.loc[df['List Price'].idxmax()]
print("\nProduct with the highest standard cost:")
print(highest_standard_cost_product)
print("\nProduct with the highest list price:")
print(highest_list_price_product)

# 5. Location Analysis
# Location with the highest quantity of products in stock
location_counts = df['Location'].value_counts()
highest_quantity_location = location_counts.idxmax()
print("\nLocation with the highest quantity of products in stock:", highest_quantity_location)

# 6. Insights and Reporting
# Summary of findings and insights
print("\nSummary of Findings:")
print("1. The dataset contains", len(df), "products.")
print("2. The category with the highest number of products is:", highest_category)
print("3. The average standard cost across all products is:", avg_standard_cost)
print("4. The average list price across all products is:", avg_list_price)
print("5. The location with the highest quantity of products in stock is:", highest_quantity_location)

# You can create a report or presentation using the findings and visualizations generated.
```
